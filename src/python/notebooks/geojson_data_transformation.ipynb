### Création des zones ###

```python
import json
```

```python
with open(f"{DATA_DIR}/gold/zones.json", "w") as file:
    empty_zones = []
    for i in range(100):
        zone = {}
        zone["id"] = i + 1
        zone["departements"] = []
        zone["parcs"] = []
        empty_zones.append(zone)

    json.dump(empty_zones, file, indent=2)
```
